```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def bar_plot(variable):
  var =data[variable]
  varValue = var.value_counts()
  plt.figure(figsize=(15,3))
  plt.bar(varValue.index, varValue,color=['#00008b','#00e5ee','#cd1076','#FF0000','#FF0000','#ADD8E6','#FFFF00','#FFC0CB','#808000','#808000','#A52A2A'])
  plt.xticks(varValue.index, varValue.index.values)
  plt.ylabel("Frequency")
  plt.title(variable)
  plt.show()
  print("{}: \n {}".format(variable,varValue))
```

```python
categoryc=['Age group']
for c in categoryc:
   bar_plot(c)
```
